## 5. Check and deal with `NaN` values.


```python
# check the Null values in the df
data.isna().sum() 
```

```python
# check columns 'state' Null 
data[data['state'].isna()==True]
```

```python
#check overall response
data['state'].value_counts()
```

```python
# fill the missing values in 'state' with 'california' #Reason: highest number of counts

data['state'] = data['state'].fillna('California')
data['state'].value_counts()
```

```python
data.isna().sum()  # check the number of Null in customer_lifetime_value 
```

```python
# check the values in 'response'
data['response'].value_counts()
```

```python
# refill with 'No' as it is the larger number
data['response'] = data['response'].fillna('No')
data['response'].value_counts()
```

```python
data['months_since_last_claim'].value_counts()
```

```python
#first check the mean
data['months_since_last_claim'].mean()
```

```python
#give a value to the mean
claim_mean = data['months_since_last_claim'].mean()
data['months_since_last_claim'] = data['months_since_last_claim'].fillna(claim_mean)
```

```python
data['months_since_last_claim'].value_counts()
```

```python
# check the number of values
data['number_of_open_complaints'].value_counts()
```

```python
#Check mean
data['number_of_open_complaints'].mean()
```

```python
#Since the mean is very close to 0.0, replace the value to 0.0
data['number_of_open_complaints'] = data['number_of_open_complaints'].fillna(0.0)
data['number_of_open_complaints'].value_counts()
```

```python
# check columns 'state' Null 
data.isna().any()
```

```python
data['vehicle_class'].value_counts()
```

```python
data['vehicle_class'] = data['vehicle_class'].fillna('Four-Door Car')
```

```python
data['vehicle_size'].value_counts()
```

```python
data['vehicle_size'] = data['vehicle_size'].fillna('Medsize')
```

```python
data['vehicle_type'].value_counts()
```

```python
data['vehicle_type'] = data['vehicle_type'].fillna('A')
```
